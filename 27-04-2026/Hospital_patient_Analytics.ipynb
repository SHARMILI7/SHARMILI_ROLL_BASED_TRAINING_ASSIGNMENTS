```python
from pyspark.sql.functions import when
patients=patients.withColumn("age_group",when(patients.age<30,"Young").when((patients.age>=30)&(patients.age<=45),"Adult").otherwise("Senior"))
patients.show()
```

```python
patients=patients.withColumn("hospital_name",lit("BotCampus Hospital"))
patients.show()
```

```python
doctors=doctors.withColumn("fee_with_tax",doctors.consultation_fee*1.18)
doctors.show()
```

```python
bills=bills.withColumn("bill_with_tax",bills.bill_amount*1.18)
bills.show()
```

```python
bills=bills.withColumn("bill_in_thousands",bills.bill_amount/1000)
bills.show()
```

```python
patients=patients.withColumn("country",lit("India"))
patients.show()
```

```python
from pyspark.sql.functions import concat,col
doctors=doctors.withColumn("doctor_label",concat(col("consultant_name"),lit("-"),col("specialization")))
doctors.show()
```

```python
patients=patients.withColumn("patient_label",concat(col("patient_name"),lit("-"),col("city")))
patients.show()
```

```python
bills=bills.withColumn("high_bill_flag",when(bills.bill_amount>1500,"Yes").otherwise("No"))
bills.show()
```

```python
patients=patients.withColumn("senior_patient_flag",when(patients.age>40,"Yes").otherwise("No"))
patients.show()
```

```python
patients=patients.withColumn("age_category",when(patients.age<30,"Young").when((patients.age>=30)&(patients.age<=45),"Adult").otherwise("Senior"))
patients.show()
```

```python
bills=bills.withColumn("bill_category",when(bills.bill_amount>1500,"High").when(bills.bill_amount>=1000,"Mediuim").otherwise("Low"))
```

```python
doctors=doctors.withColumn("fee_category",when(doctors.consultation_fee>1500,"Premiuim").when(doctors.consultation_fee>=1000,"Standara").otherwise("Basic"))
```

```python
appointments=appointments.withColumn("priority",when(appointments.status=="Pending","High").when(appointments.status=="Completed","Normal").otherwise("Low"))
```

```python
patients=patients.withColumn("city_zone",when(patients.city.isin("Hyderabad","Bangalore","Chennai"),"South").otherwise("Metro"))
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number,rank,dense_rank
```

```python
window=Window.partitionBy("city").orderBy(col("age").desc())
patients.withColumn("rank",rank().over(window)).show()
```

```python
patients.withColumn("row_num",row_number().over(window)).filter("row_num=1").show()
```

```python
from pyarrow.util import doc
window2=Window.partitionBy("specialization").orderBy(col("consultation_fee").desc())
doctors.withColumn("rank",dense_rank().over(window2)).show()
```

```python
doctors.withColumn("rank",dense_rank().over(window2)).filter("rank=1").show()
```

```python
patients.withColumn("rank",rank().over(window)).filter("rank<=2").show()
```

```python
window3=Window.orderBy(col("consultation_fee").desc())
doctors.withColumn("rank",rank().over(window3)).show()
```

```python
city_df=patients.groupBy("city").count()
window4=Window.orderBy(col("count").desc())
city_df.withColumn("rank",rank().over(window4)).show()
```

```python
doc_df=appointments.groupBy("doctor_id").count()
window5=Window.orderBy(col("count").desc())
doc_df.withColumn("rank",rank().over(window5)).show()
```

```python
window6=Window.partitionBy("payment_mode").orderBy("bill_amount")
bills.withColumn("running_total",sum("bill_amount").over(window6)).show()
```

```python
window7=Window.partitionBy("doctor_id").orderBy("appointment_date")
appointments.withColumn("running_count",count("*").over(window7)).show()
```
