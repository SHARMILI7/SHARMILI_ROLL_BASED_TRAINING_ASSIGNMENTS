```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank
df=customers.join(accounts,"customer_id")
df.withColumn("rank",rank().over(Window.partitionBy("city").orderBy(col("balance").desc()))).show()
```

```python
from pyspark.sql.functions import row_number
df.withColumn("row_num",row_number().over(Window.partitionBy("city").orderBy(col("balance").desc()))).filter(col("row_num")==1).show()
```

```python
from pyspark.sql.functions import dense_rank
transactions.withColumn("dense_rank",dense_rank().over(Window.partitionBy("txn_type").orderBy(col("amount").desc()))).show()
```

```python
transactions.withColumn("rank",row_number().over(Window.partitionBy("account_id").orderBy(col("amount").desc()))).filter(col("rank")<=2).show()
```

```python
from pyspark.sql.functions import sum
transactions.withColumn("running_total",sum("amount").over(Window.partitionBy("account_id").orderBy("txn_date").rowsBetween(Window.unboundedPreceding,0))).show()
```

```python
from pyspark.sql.functions import sum,rank
from pyspark.sql.window import Window
df=accounts.groupBy("branch").agg(sum("balance").alias("total_balance"))
df.withColumn("rank",rank().over(Window.orderBy(col("total_balance").desc()))).show()
```

```python
df=customers.join(accounts,"customer_id").groupBy("city").agg(sum("balance").alias("total_balance"))
df.withColumn("rank",rank().over(Window.orderBy(col("total_balance").desc()))).show()
```

```python
df=customers.join(accounts,"customer_id").join(transactions,"account_id").groupBy("name").agg(sum("amount").alias("total_amount"))
df.withColumn("rank",rank().over(Window.orderBy(col("total_amount").desc()))).show()
```

```python
from pyspark.sql.functions import row_number
df=customers.join(accounts,"customer_id").join(transactions,"account_id")
df.withColumn("row_num",row_number().over(Window.partitionBy("city").orderBy(col("amount").desc()))).filter(col("row_num")==1).show()
```

```python
df=customers.join(accounts,"customer_id").join(branches,"branch")
df.withColumn("row_num",row_number().over(Window.partitionBy("region").orderBy(col("balance").desc()))).filter(col("row_num")==1).show()
```
