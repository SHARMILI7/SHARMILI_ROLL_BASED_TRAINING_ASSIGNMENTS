```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,count
spark=SparkSession.builder.appName("LoginAnalysis").getOrCreate()
df=spark.read.text("logins.txt")
df=df.withColumnRenamed("value","name")
```

```python
df.show()
total_logins=df.count()
print(total_logins)
unique_users=df.select("name").distinct()
unique_users.show()
user_counts=df.groupBy("name").count()
user_counts.show()
top_users=user_counts.orderBy(col("count").desc()).limit(3)
top_users.show()
active_users=user_counts.filter(col("count")>4)
active_users.show()
result_dict={row["name"]:row["count"] for row in user_counts.collect()}
print(result_dict)
```

```python
from google.colab import files
uploaded=files.upload()
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,avg,count,max,min
spark=SparkSession.builder.appName("EmployeeAnalaysis").getOrCreate()
df=spark.read.csv("employees.csv",header=True,inferSchema=True)
df.show()
```

```python
total_emp=df.count()
print(total_emp)
df.filter(col("department")=="IT").show()
df.filter(col("salary")>75000).show()
df.select(avg("salary")).show()
df.orderBy(col("salary").desc()).limit(1).show()
df.orderBy(col("salary")).limit(1).show()
df.groupBy("department").count().show()
df.groupBy("department").agg(avg("salary").alias("avg_salary")).show()
df.groupBy("city").count().show()
df.orderBy(col("salary").desc()).limit(5).show()
df.filter((col("city")=="Hyderabad") & (col("salary")>70000)).show()
```

```python
from google.colab import files
uploaded=files.upload()
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,sum,avg
spark=SparkSession.builder.appName("SalesAnalysis").getOrCreate()
df=spark.read.csv("sales.csv",header=True,inferSchema=True)
df.show()
```

```python
df=df.withColumn("revenue",col("quantity")*col("price"))
df.show()
total_revenue=df.select(sum("revenue"))
total_revenue.show()
revenue_per_product=df.groupBy("product").agg(sum("revenue").alias("total_revenue"))
revenue_per_product.show()
df.groupBy("product").agg(sum("quantity").alias("total_quantity")).show()
df.groupBy("product").agg(sum("quantity").alias("total_quantity")).orderBy(col("total_quantity").desc()).show()
df.groupBy("emp_id").agg(sum("revenue").alias("total_revenue")).orderBy(col("total_revenue").desc()).limit(1).show()
df.select(avg("revenue")).show()
revenue_per_product.filter(col("total_revenue")>100000).show()
```

```python
from google.colab import files
uploaded=files.upload()
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,sum
spark=SparkSession.builder.appName("OrderAnalysis").getOrCreate()
df=spark.read.option("multiLine",True).json("orders.json")
from pyspark.sql.functions import explode
df=df.select(explode("orders").alias("order")).select("order.*")
df.show()
```

```python
df.show()
print(df.count())
df.select(sum("amount")).show()
customer_spending=df.groupBy("customer").agg(sum("amount").alias("total_spent"))
customer_spending.show()
customer_spending.orderBy(col("total_spent").desc()).limit(1).show()
df.groupBy("product").agg(sum("amount").alias("total_sales")).show()
df.filter(col("city")=="Hyderabad").select("customer").distinct().show()
df.filter(col("amount")>10000).show()
df.groupBy("city").count().show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,sum
spark=SparkSession.builder.appName("FinalChallenge").getOrCreate()
emp_df=spark.read.csv("employees.csv",header=True,inferSchema=True)
sales_df=spark.read.csv("sales.csv",header=True,inferSchema=True)
```

```python
df=emp_df.join(sales_df,on="emp_id",how="inner")
df.show()
df=df.withColumn("revenue",col("quantity")*col("price"))
df.show()
emp_revenue=df.groupBy("emp_id","name").agg(sum("revenue").alias("total_revenue"))
emp_revenue.show()
top5=emp_revenue.orderBy(col("total_revenue").desc()).limit(5)
top5.show()
dept_revenue=df.groupBy("department").agg(sum("revenue").alias("total_revenue"))
dept_revenue.orderBy(col("total_revenue").desc()).limit(1).show()
top5.write.mode("overwrite").csv("final_sales_repot.csv",header=True)
```
