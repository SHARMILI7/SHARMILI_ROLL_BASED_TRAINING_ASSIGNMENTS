```python
products_df.write.format("parquet").mode("overwrite").save("/FileStore/products_parquet")
```

```python
parquet_df=spark.read.format("parquet").load("/FileStore/products_parquet")
display(parquet_df)
```

```python
parquet_df.write.format("delta").mode("overwrite").saveAsTable("products_delta_table")
```

```python
spark.read.table("products_delta_table").show()
```

Delta supports ACID transactions,time travel,update,delete and merge while Parquet is mainly storage format without these features.

```python
spark.sql("update products_delta_table set warehouse_city='Chennai' where product_id=101")
```

Delta supports updates better because it allows direct row-level modifications without rewriting the full dataset unlike raw Parquet.
