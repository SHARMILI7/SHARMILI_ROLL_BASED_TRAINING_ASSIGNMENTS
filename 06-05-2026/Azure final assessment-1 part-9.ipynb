Delta table supports ACID transactions,time travel,merge,update and delete operations.
Parquet table does not support these advanced features directly.

```python
spark.sql("update patients_delta_table set city='Chennai' where patient_id=1001")
```

Update is easier on Delta because Delta Lake supports direct update and delete operations.
In raw Parquet,we usually need to rewrite the entire dataset after modifications
