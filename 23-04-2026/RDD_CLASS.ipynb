```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("RDD BASICS").getOrCreate()
sc=spark.sparkContext
print("spark ready")
```

```python
numbers=[10,20,30,40,50]
rdd=sc.parallelize(numbers)
print(rdd)
```

```python
rdd.collect()
```

```python
rdd=sc.parallelize([1,2,3,4,5])
mapped_rdd=rdd.map(lambda x:x*2)
fliter_rdd=mapped_rdd.filter(lambda x:x>5)
print("Nothing to be executed")
```

Here if u see we won't get any kind of output here.Because rdd just creates here not output .output will be show only when we use collect()..this is called as lazy evaluation

```python
fliter_rdd.collect()
```

```python
rdd=sc.parallelize(["rahul","sneha","arjun"])
upper_rdd=rdd.map(lambda x:x.upper())
upper_rdd.collect()
```

```python
rdd=sc.parallelize([1,2,3,4,5])
squared_rdd=rdd.map(lambda x:x*x)
squared_rdd.collect()
```

```python
rdd=sc.parallelize([10,15,20,30,25])
even_rdd=rdd.filter(lambda x:x%2==0)
even_rdd.collect()
```

```python
rdd=sc.parallelize(["apple","banana","mango","kiwi"])
long_words=rdd.filter(lambda x:len(x)>5)
long_words.collect()
```

```python
lines=sc.parallelize([
    "python spark python",
    "spark hadoop.spark",
    "python hadoop"
])
words_count=(
    lines
    .flatMap(lambda line:line.split(" "))
    .map(lambda word:(word,1))
    .reduceByKey(lambda a,b:a+b)

)
words_count.collect()
```

```python
employees=sc.parallelize([
    ("Arjun",700000),
    ("Rahul",80000),
    ("Priya",58000),
    ("Arun",75000),
    ("Mana",60000)
])
high_salary=employees.filter(lambda x:x[1]>65000)
high_salary.collect()
```
