````python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("Exercise 1").getOrCreate()
````

### Exercise 1

Create a DataFrame with:

```
Name
Salary
Department
```

Add 5 employees.

---

````python
data=[("Karan",25000,"IT"),
      ("Rishika",45000,"HR"),
      ("Deva",75000,"Sales"),
      ("Prithvi",90000,"HR"),
      ("Richard",100000,"IT")
      ]
column=["Name","Salary","Department"]
df=spark.createDataFrame(data,column)
df.show()
````

### Exercise 2

Show only **Name and Salary** columns.

---

````python
df.select("Name","Salary").show()
````

### Exercise 3

Filter employees with **salary greater than 50000**.

---

````python
df.filter(df.Salary>50000).show()
````

### Exercise 4

Count employees per department.

---

````python
df.groupBy("Department").count().show()
````

### Exercise 5

Sort employees by salary.

---

````python
df.orderBy("Salary").show()
````

````python
df.orderBy(df.Salary.desc()).show()
````
