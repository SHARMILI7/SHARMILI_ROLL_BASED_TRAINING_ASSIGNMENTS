```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
def categorize_sales(amount):
  if amount>50000:
    return "Premium"
  elif amount>=20000:
    return "Medium"
  else:
    return "Small"
category_udf=udf(categorize_sales,StringType())
sales=sales.withColumn("sale_category",category_udf(col("amount")))
sales.show()
```

```python
import ipywidgets as widgets
departments_list=[row["department"] for row in employees.select("department").distinct().collect()]
dept_widget=widgets.Dropdown(options=departments_list,description='Department:')
display(dept_widget)
```

```python
salary_slider=widgets.IntSlider(min=50000,max=80000,step=1000,description='Salary:')
display(salary_slider)
```

```python
employees.filter((col("department")==dept_widget.value) & (col("salary")>=salary_slider.value)).show()
```

```python
from pyspark.sql.functions import avg,sum
top_employee=employees.orderBy(col("salary").desc()).first()
avg_salary=employees.select(avg("salary")).first()[0]
dept_count=employees.groupBy("department").count().collect()
total_sales=sales.select(sum("amount")).first()[0]
best_sales_emp=sales.groupBy("emp_id").agg(sum("amount").alias("total")).orderBy(col("total").desc()).first()
```

```python
with open("final_report.txt","w") as f:
  f.write("E-Commerce Report\n\n")
  f.write(f"Top Employee:{top_employee['name']}\n")
  f.write(f"Average Salary:{avg_salary}\n")
  f.write(f"Total Sales:{total_sales}\n")
  f.write(f"Best Sales Employee ID :{best_sales_emp['emp_id']}\n\n")
  f.write("Department Employee Count:\n")
  for row in dept_count:
    f.write(f"{row['department']}:{row['count']}\n")
```
